```python
import numpy as np
```

```python
T = 1.0
F = 0.0

#creating data for OR operation

def get_OR_data():
    X = [
        [F, F],
        [F, T],
        [T, F],
        [T, T]
    ]

    Y = [
        [F],
        [T],
        [T],
        [T]
    ]
    return X, Y

X, Y = get_OR_data()
```

The second function is divided into four stages:
* Deciding weight and bias
* Deciding the activation function
* Deciding the loss function
* Deciding the optimization algorithm

```python
class MyPerceptron:
    def __init__(self, learning_rate=0.1, n_iters=1000):
        self.lr = learning_rate
        self.epochs = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, Y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        #training
        for epoch in range(self.epochs):
            for i in range(X.shape[0]):
                Y_pred = self.step_function(np.dot(X[i], self.weights) + self.bias) #activation function
                mae = Y[i] - Y_pred #loss function

                self.weights += self.lr * mae * X[i]
                self.bias += self.lr * mae

    def step_function(self, activation):
        return 1 if activation >= 0 else 0
    
    def predict(self, X):
        y_preds = []
        for x in X:
            linear_output = np.dot(x, self.weights) + self.bias
            y_preds.append(self.step_function(linear_output))
        return np.array(y_preds)
```

```python
X = np.array(X)
Y = np.array(Y)
Y = Y.flatten()

clf = MyPerceptron()
clf.fit(X, Y)

X_test = [
    [F, F],
    [T, F],
    [F, T],
    [F, F]
]

Y_test = [
    [F],
    [T],
    [T],
    [F]
]

X_test = np.array(X_test)
Y_test = np.array(Y_test)
Y_predicted = clf.predict(X_test)
print(Y_predicted)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import tensorflow as tf
```

```python
T = 1.0
F = 0.0

def get_XOR_data():
    X = [
        [F, F],
        [F, T],
        [T, F],
        [T, T]
    ]

    Y = [
        [F],
        [T],
        [T],
        [F]
    ]
    return X, Y

X, Y = get_XOR_data()

X_test = [
    [T, F],
    [T, T],
    [F, T],
    [F, F]
]

Y_test = [
    [T],
    [F],
    [T],
    [F]
]
```

```python
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(np.array(X), np.array(Y), epochs=1000)
loss, accuracy = model.evaluate(np.array(X_test), np.array(Y_test), verbose=0)
print(f'Loss: {loss}, Accuracy: {accuracy}')

y_pred_prob = model.predict(np.array(X_test))
y_pred = (y_pred_prob >= 0.5).astype(int)
print(y_pred)
```
